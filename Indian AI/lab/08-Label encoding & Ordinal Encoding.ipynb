# Label encoding & Ordinal Encoding

```python
pd.set_option("display.max_rows",None)
```

Ordinal Encoding

```python
label = {"Ex":4,"Gd":3,"TA":2,"Fa":1}
```

```python
df['KitchenQual_new'] = df['KitchenQual'].map(label)
```

```python
df['KitchenQual_new'].value_counts()
```

```python
df['KitchenQual_new'].unique()
```

Label encoding

```python
from sklearn.preprocessing import LabelEncoder 
```

```python
le = LabelEncoder()
```

```python
df['RoofStyle_new'] = le.fit_transform(df['RoofStyle'])
```

```python
df.RoofStyle_new.value_counts()
```
